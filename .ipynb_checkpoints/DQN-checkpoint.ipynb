```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import random
import matplotlib.pyplot as plt
from collections import deque
```

```python
# --- 1. IMPORT THE ENVIRONMENT ---
# We use a trick to import the other .ipynb file as a library
import import_ipynb 
import RYU_ENV as env_module

print("âœ… Successfully connected to Ryu_Environment_Setup!")
```

```python
# --- 2. DEFINE THE AGENT (The Brain) ---
class DQNAgent(nn.Module):
    def __init__(self, state_dim, action_dim):
        super(DQNAgent, self).__init__()
        self.state_dim = state_dim
        self.action_dim = action_dim
        
        # The Neural Network Layers
        self.net = nn.Sequential(
            nn.Linear(state_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, action_dim)
        )
        
        self.optimizer = optim.Adam(self.net.parameters(), lr=0.001)
        self.loss_fn = nn.MSELoss()
        self.memory = deque(maxlen=2000)
        self.epsilon = 1.0  # 100% Random at start
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995

    def forward(self, x):
        return self.net(x)

    def select_action(self, state):
        # Exploration (Random)
        if random.random() < self.epsilon:
            return random.randint(0, self.action_dim - 1)
        
        # Exploitation (Brain)
        with torch.no_grad():
            state_t = torch.FloatTensor(state).unsqueeze(0)
            q_values = self.net(state_t)
            return q_values.argmax().item()

    def store_experience(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def train_step(self, batch_size=32):
        if len(self.memory) < batch_size: return 0.0
        
        batch = random.sample(self.memory, batch_size)
        states, actions, rewards, next_states, dones = zip(*batch)
        
        states = torch.FloatTensor(np.array(states))
        actions = torch.LongTensor(actions).unsqueeze(1)
        rewards = torch.FloatTensor(rewards).unsqueeze(1)
        next_states = torch.FloatTensor(np.array(next_states))
        dones = torch.FloatTensor(dones).unsqueeze(1)
        
        # Calculate Target Q (Bellman Equation)
        with torch.no_grad():
            next_q = self.net(next_states).max(1)[0].unsqueeze(1)
            target_q = rewards + (0.99 * next_q * (1 - dones))
        
        current_q = self.net(states).gather(1, actions)
        
        loss = self.loss_fn(current_q, target_q)
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
        
        return loss.item()
```

```python
# --- 3. INITIALIZE TRAINING ---
print("ðŸš€ Initializing Training System...")

# A. Load the Mock Ryu Controller
# Ensure 'synthetic_traffic_16hosts.csv' is in this folder!
mock_ryu = env_module.MockRyuController("synthetic_traffic_16hosts.csv")

# B. Determine Input Size (Run once to check)
dummy_stats = mock_ryu.get_all_switch_stats()
# Flatten the stats to count inputs
flat_state = []
for sw in range(1, 21):
    for port in range(1, 5):
        flat_state.append(dummy_stats[sw][port])

STATE_DIM = len(flat_state)   # Should be 80 (20 switches * 4 ports)
ACTION_DIM = STATE_DIM        # Agent can modify any of the 80 ports
print(f"   -> State Size: {STATE_DIM}, Action Size: {ACTION_DIM}")

# C. Create the Agent
agent = DQNAgent(STATE_DIM, ACTION_DIM)
print("   -> Agent Online.")

# --- 4. THE MAIN LOOP ---
EPISODES = 50
history_rewards = []

print("\nStarting Training Episodes...")
for episode in range(EPISODES):
    mock_ryu.time_step = 0 # Reset Simulation Clock
    total_reward = 0
    
    # Get initial state
    raw_stats = mock_ryu.get_all_switch_stats()
    state = []
    for sw in range(1, 21):
        for port in range(1, 5):
            state.append(raw_stats[sw][port])
    state = np.array(state)
    
    for step in range(100): # 100 decisions per episode
        # 1. Agent Decides
        action = agent.select_action(state)
        
        # 2. Decode Action
        target_switch = (action // 4) + 1
        target_port = (action % 4) + 1
        
        # 3. Apply to Environment
        mock_ryu.mod_flow(target_switch, target_port)
        mock_ryu.tick() # Move time forward
        
        # 4. Get New State
        raw_stats_next = mock_ryu.get_all_switch_stats()
        next_state = []
        for sw in range(1, 21):
            for port in range(1, 5):
                next_state.append(raw_stats_next[sw][port])
        next_state = np.array(next_state)
        
        # 5. Calculate Reward (Minimize Congestion)
        max_util = np.max(next_state)
        reward = 1.0 / (max_util + 1.0) # Higher reward for lower traffic
        
        # 6. Train
        done = (step == 99)
        agent.store_experience(state, action, reward, next_state, done)
        agent.train_step()
        
        state = next_state
        total_reward += reward
        
    # Decay Randomness
    agent.epsilon = max(agent.epsilon_min, agent.epsilon * agent.epsilon_decay)
    history_rewards.append(total_reward)
    
    if (episode+1) % 5 == 0:
        print(f"Episode {episode+1}: Total Reward = {total_reward:.2f}, Epsilon = {agent.epsilon:.2f}")

```

```python
# --- 5. SAVE & PLOT ---
torch.save(agent.state_dict(), "ryu_dqn_brain.pth")
print("\nâœ… Training Complete. Brain saved as 'ryu_dqn_brain.pth'")

plt.plot(history_rewards)
plt.title("Agent Learning Progress")
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.show()
```
