```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import random
from collections import deque

# --- 2. DEFINE THE AGENT (The Brain) ---
class DQNAgent(nn.Module):
    def __init__(self, state_dim, action_dim):
        super(DQNAgent, self).__init__()
        self.state_dim = state_dim
        self.action_dim = action_dim
        
        # The Neural Network Layers (Input -> Hidden -> Output)
        self.net = nn.Sequential(
            nn.Linear(state_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, action_dim) # Output: Q-value for each possible Action
        )
        
        self.optimizer = optim.Adam(self.net.parameters(), lr=0.001)
        self.loss_fn = nn.MSELoss()
        self.memory = deque(maxlen=5000) # Larger memory for stability
        
        # Hyperparameters
        self.epsilon = 1.0       # 100% Random at start (Exploration)
        self.epsilon_min = 0.05  # Keep 5% randomness so it keeps learning
        self.epsilon_decay = 0.995
        self.gamma = 0.95        # Discount factor (Future rewards matter)

    def forward(self, x):
        return self.net(x)

    def select_action(self, state):
        # Exploration (Random Guess)
        if random.random() < self.epsilon:
            return random.randint(0, self.action_dim - 1)
        
        # Exploitation (Use Brain)
        with torch.no_grad():
            state_t = torch.FloatTensor(state).unsqueeze(0)
            q_values = self.net(state_t)
            return q_values.argmax().item()

    def store_experience(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def train_step(self, batch_size=64):
        if len(self.memory) < batch_size: return 0.0
        
        # Sample random batch from memory
        batch = random.sample(self.memory, batch_size)
        states, actions, rewards, next_states, dones = zip(*batch)
        
        # Convert to Tensors
        states = torch.FloatTensor(np.array(states))
        actions = torch.LongTensor(actions).unsqueeze(1)
        rewards = torch.FloatTensor(rewards).unsqueeze(1)
        next_states = torch.FloatTensor(np.array(next_states))
        dones = torch.FloatTensor(dones).unsqueeze(1)
        
        # 1. Calculate Target Q (Bellman Equation)
        with torch.no_grad():
            # What is the best possible value of the NEXT state?
            next_q = self.net(next_states).max(1)[0].unsqueeze(1)
            target_q = rewards + (self.gamma * next_q * (1 - dones))
        
        # 2. Get Current Q (Prediction)
        current_q = self.net(states).gather(1, actions)
        
        # 3. Backpropagation (Learn)
        loss = self.loss_fn(current_q, target_q)
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
        
        # Decay Epsilon (Get smarter over time)
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay
            
        return loss.item()
```

```python
class NetworkEnv:
    def __init__(self, controller):
        self.mock = controller
        # Map Action ID (0-15) to (Switch_ID, Out_Port)
        # We have 8 Edge Switches: [7, 8, 11, 12, 15, 16, 19, 20]
        # Each has 2 Uplinks: Port 1, Port 2
        self.action_map = {}
        idx = 0
        for sw in self.mock.edge_switches:
            self.action_map[idx] = (sw, 1) # Action 0, 2, 4... -> Port 1 (Path A)
            idx += 1
            self.action_map[idx] = (sw, 2) # Action 1, 3, 5... -> Port 2 (Path B)
            idx += 1
            
        self.state_dim = 4 + (8 * 2) # 4 Core Links + 16 Edge Uplinks
        self.action_dim = len(self.action_map)

    def get_state(self):
        """Flatten the network stats into a single array for the AI"""
        util, _ = self.mock.get_stats()
        state = []
        
        # 1. Core Switch Loads (The most important info)
        # Core 1-4, Port 1 is the main downlink
        for c in range(1, 5):
            state.append(util[c][1]) 
            
        # 2. Edge Switch Uplink Loads
        for sw in self.mock.edge_switches:
            state.append(util[sw][1]) # Uplink 1
            state.append(util[sw][2]) # Uplink 2
            
        return np.array(state, dtype=np.float32)

    def step(self, action_idx):
        """Execute one action and calculate reward"""
        # 1. Decode Action
        target_sw, target_port = self.action_map[action_idx]
        
        # 2. Apply to Network
        self.mock.mod_flow(target_sw, target_port)
        
        # 3. Tick Clock (Physics Engine)
        self.mock.tick()
        
        # 4. Get New State
        next_state = self.get_state()
        util, _ = self.mock.get_stats()
        
        # 5. CALCULATE REWARD (Crucial!)
        # Goal: Minimize Max Utilization in the Core
        # If Core is 150% overloaded, Reward should be low (Negative)
        # If Core is 50% loaded, Reward should be high (Near Zero)
        
        core_loads = [util[c][1] for c in range(1, 5)]
        max_load = max(core_loads)
        mean_load = np.mean(core_loads)
        
        # Reward Formula:
        # Penalize Overload heavily.
        # Slight penalty for imbalance.
        reward = -1.0 * max_load 
        
        if max_load > 1.0: # If congestion exists
            reward -= 5.0 # BIG Punishment for Red Lights
            
        done = False # Continuous task
        
        return next_state, reward, done
```

```python
# --- 0. IMPORTS (CRITICAL) ---
import import_ipynb 
import RYU_ENV as env_module  # We nickname the file 'env_module'

# --- CONFIGURATION ---
EPISODES = 200
BATCH_SIZE = 32
TRAFFIC_FILE = "Dataset/Synthesis/synthetic_traffic_16hosts_FINAL1.csv"

# 1. Init System
print("üöÄ STARTING TRAINING SESSION...")

# ‚úÖ CORRECT FIX: Access the 'MockRyuController' class INSIDE the file
mock_controller = env_module.MockRyuController(TRAFFIC_FILE)

# Initialize Wrapper and Agent (Assumes these classes are defined in THIS notebook)
env = NetworkEnv(mock_controller)
agent = DQNAgent(state_dim=env.state_dim, action_dim=env.action_dim)

rewards_history = []
loss_history = []

# 2. Training Loop
for episode in range(EPISODES):
    state = env.get_state()
    total_reward = 0
    
    # Run for 50 steps per episode (Simulation Time)
    for step in range(50):
        # A. Select Action
        action = agent.select_action(state)
        
        # B. Apply Action
        next_state, reward, done = env.step(action)
        
        # C. Store Memory
        agent.store_experience(state, action, reward, next_state, done)
        
        # D. Learn
        loss = agent.train_step(BATCH_SIZE)
        
        state = next_state
        total_reward += reward
        if loss is not None:
             loss_history.append(loss)

    rewards_history.append(total_reward)
    
    # Print Progress every 10 episodes
    if episode % 10 == 0:
        avg_loss = np.mean(loss_history[-50:]) if loss_history else 0
        print(f"   Episode {episode:>3} | Reward: {total_reward:>6.1f} | Epsilon: {agent.epsilon:.2f} | Loss: {avg_loss:.4f}")

# 3. Save the Brain
import os
os.makedirs("models", exist_ok=True) # Ensure folder exists
torch.save(agent.state_dict(), "models/dqn_traffic_agent.pth")
print("\nüíæ Saved trained model to 'models/dqn_traffic_agent.pth'")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch

# --- CONFIGURATION ---
TEST_DURATION = 100  # Longer test to see stability
TRAFFIC_FILE = "Dataset/Synthesis/synthetic_traffic_16hosts_FINAL1.csv"
MODEL_PATH = "models/dqn_traffic_agent.pth"

print("‚öîÔ∏è STARTING FINAL BATTLE: Trained Agent vs. Standard Dijkstra...")

# --- 1. SETUP THE ARENA ---
# We use the existing classes you defined
# Make sure imports are correct
import RYU_ENV as env_module
mock_controller = env_module.MockRyuController(TRAFFIC_FILE)
env = NetworkEnv(mock_controller)

# Load the Trained Brain
trained_agent = DQNAgent(state_dim=env.state_dim, action_dim=env.action_dim)
try:
    trained_agent.load_state_dict(torch.load(MODEL_PATH))
    trained_agent.epsilon = 0.0 # üö® CRITICAL: Turn off randomness
    trained_agent.eval()
    print("‚úÖ Loaded Trained Model.")
except FileNotFoundError:
    print("‚ùå Error: Model not found. Please train it first!")
    exit()

# --- 2. DEFINE THE RUNNERS ---

def run_standard_algorithm():
    """
    Simulates Standard Routing (Dijkstra / ECMP).
    The Agent does NOTHING. Traffic follows the default hash logic.
    """
    print("   üõ£Ô∏è  Running Standard Dijkstra (Baseline)...")
    env.mock.time_step = 0 
    env.mock._reset_stats()
    env.mock.flow_table = {} # Clear any old rules
    
    congestion_history = []
    
    for step in range(TEST_DURATION):
        # 1. Physics Engine (Get Stats)
        util, _ = env.mock.get_stats()
        
        # 2. Record Performance (Max Core Link Load)
        # We check Core Switches 1-4 (Index 1-4 in util)
        # Port 1 is the main downlink
        core_loads = [util[c][1] for c in range(1, 5)]
        max_load = max(core_loads) * 100 # Convert to %
        congestion_history.append(max_load)
        
        # 3. Tick Clock (No Action Taken)
        env.mock.tick()
        
    return congestion_history

def run_smart_agent():
    """
    Simulates the RL Agent.
    The Agent actively modifies flow tables to reduce congestion.
    """
    print("   üß† Running RL Agent (Smart)...")
    env.mock.time_step = 0 
    env.mock._reset_stats()
    env.mock.flow_table = {} # Clear any old rules
    
    state = env.get_state()
    congestion_history = []
    
    for step in range(TEST_DURATION):
        # 1. AI Makes Decision
        action = trained_agent.select_action(state)
        
        # 2. Apply Decision (Reroute Traffic)
        next_state, _, _ = env.step(action)
        
        # 3. Record Performance
        # We check Core Switches 1-4
        util, _ = env.mock.get_stats()
        core_loads = [util[c][1] for c in range(1, 5)]
        max_load = max(core_loads) * 100
        congestion_history.append(max_load)
        
        state = next_state
        
    return congestion_history

# --- 3. RUN THE RACE ---
history_dijkstra = run_standard_algorithm()
history_rl = run_smart_agent()

# --- 4. THE VERDICT (Visual Proof) ---
print("\nüìä FINAL REPORT:")
avg_base = np.mean(history_dijkstra)
avg_smart = np.mean(history_rl)
peak_base = np.max(history_dijkstra)
peak_smart = np.max(history_rl)

improvement = ((avg_base - avg_smart) / avg_base) * 100

print(f"   Avg Load (Dijkstra): {avg_base:.1f}%")
print(f"   Avg Load (RL Agent): {avg_smart:.1f}%")
print(f"   ---------------------------")
print(f"   Peak Load (Dijkstra): {peak_base:.1f}%")
print(f"   Peak Load (RL Agent): {peak_smart:.1f}%")
print(f"   ---------------------------")
print(f"   üöÄ EFFICIENCY GAIN: {improvement:.1f}%")

# Plotting
plt.figure(figsize=(12, 6))

# Plot Baseline (Red Area)
plt.fill_between(range(TEST_DURATION), history_dijkstra, color='red', alpha=0.1)
plt.plot(history_dijkstra, color='red', linestyle='--', linewidth=1.5, label="Standard (Dijkstra)")

# Plot AI (Blue Line)
plt.plot(history_rl, color='blue', linewidth=2.5, label="RL Agent (DQN)")

plt.axhline(y=100, color='black', linestyle='-', linewidth=1, label="Link Capacity (100%)")

plt.title(f"Congestion Control: RL Agent vs Standard Routing", fontsize=14)
plt.ylabel("Max Core Utilization (%)", fontsize=12)
plt.xlabel("Time (Simulation Steps)", fontsize=12)
plt.legend(loc="upper right")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
